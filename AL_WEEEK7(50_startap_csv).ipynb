```python
import pandas as pd 
url = '/content/Startups.csv'
data = pd.read_csv(url)

data.head()
```

```python
# khổ phân bố dữ liệu 
import matplotlib.pyplot as plt 
data.hist()   
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
#màu càng đậm giá trị càng thấp 
import seaborn as sns 
sns.heatmap(correlation, annot=True ) 
```

```python
sns.pairplot(data)  
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['R&D Spend']), 
          y=data['Profit'])

print('Hệ số intercept là: ', model.intercept_)
print('Hệ số coeffient là: ', model.coef_)

```

```python
#độ chính xác 
model.score(X=pd.DataFrame(data['R&D Spend']), 
          y=data['Profit']) 
```

```python
# dự đoán 
y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
print(y_pred)
```

```python
# vẽ đồ thị 
data.plot(kind='scatter', x='R&D Spend', y='Profit', figsize=(9,9),
          color = 'black', xlim=(0,200010))
                                   
plt.plot(data['R&D Spend'], y_pred, color='green')
```

```python
#bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'], data['R&D Spend']**1, data['R&D Spend']**2]).T
y=data['Profit']

model.fit(X,y=data['Profit'])
print('Intecept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y))


```

```python
import numpy as np
range = np.arange(0, 200000, 10)
X_range=pd.DataFrame([range,range**1, range**2]).T 
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x='R&D Spend', 
          y = 'Profit', figsize = (9,9), color = 'black',
          xlim = (0,200010))
plt.plot(range, y_pred, color = 'red')
```
