```python
class_name = [ "T-shir/top" ,"Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
import matplotlib.pyplot as plt 
for i in range(25):
  plt.subplot(5,5, i+1)
  plt.tight_layout()
  plt.imshow(x_train[i], cmap = "gray", interpolation = "none")
  plt.title("{}".format(class_name[y_train[i]]))
  plt.xticks([])
  plt.yticks([])
```

```python

from keras.models import Sequential #Sequnetial: la mang ANN, lop nay nos ke tiep nhau (mo ta mang ANN theo tung lop)
from keras.layers import Dense  # Dense = full_connectied : ket noi day du theo cac te bao 

model = Sequential()
model.add(Dense(512, activation = 'relu', input_shape = (28*28,), name = "layer_1"))  # them mot lop: add, them mot full_connectied: Dense, 512: so noron duoc them 
model.add(Dense(256, activation = 'relu', input_shape = (28*28,), name = "layer_2")) 
model.add(Dense(128, activation = 'relu', input_shape = (28*28,), name = "layer_3")) 
model.add(Dense(10, activation = 'softmax', name = "Layer")) 

model.summary() 
     
```

```python
X = x_test   # luu du lieu
Y = y_test    # luu du lieu
```

```python
x_train = x_train.reshape(60000, 28*28)    # tuwf 60000 hangf thanh 1 hang, 28*28 thanh 1 cot, du lieu cua train
x_train = x_train.astype('float32')/255   # chuyen thanh kieu thuc tu 0 den 1
x_test = x_test.reshape(10000,28*28)       # du lieu cua test
x_test = x_test.astype('float32')/255

x_train.shape
```

```python
x_test.shape
```

```python
from keras.utils import to_categorical        # categorical chuyen 1 bit thanh n bits 
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model.compile(optimizer = 'rmsprop', loss = 'categorical_crossentropy',metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 30, batch_size = 128)    #batch_size: số dữ liệu mỗi lần học 
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Mat mat: ", test_loss)
print("Do chinh xac: ", test_acc)
```

```python
model.save("fashion_minist.h5")
```
