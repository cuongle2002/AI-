```python
import pandas as pd 
url = '/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt 
data.hist()   # khổ phân bố dữ liệu 
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
import seaborn as sns 
sns.heatmap(correlation, annot=True ) #màu càng đậm giá trị càng thấp 
```

```python
sns.pairplot(data) 
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']), 
          y=data['mpg'])
print('Hệ số intercept là: ', model.intercept_)
print('Hệ số coeffient là: ', model.coef_)

#pt hồi quy 
 # (mpg = -5,344wt+37,285)
```

```python
#độ chính xác 
model.score(X=pd.DataFrame(data['wt']), 
          y=data['mpg']) 
```

```python
# dự đoán 
y_pred = model.predict(X = pd.DataFrame(data['wt']))
print(y_pred)
```

```python
# vẽ đồ thị 
data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9),
          color = 'black', xlim=(0,7))
plt.plot(data['wt'], y_pred, color='yellow')
```

```python
#
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2]).T
y=data['mpg']

model.fit(X,y=data['mpg'])
print('Intecept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y))

#pt hồi quy là: 
# (-13.38wt + 1.17wt^2 + 49.93)
```

```python
import numpy as np
range = np.arange(1.5, 5.5, 0.1)
X_range=pd.DataFrame([range,range**2]).T 
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x='wt', 
          y = 'mpg', figsize = (9,9), color = 'black',
          xlim = (0,7))
plt.plot(range, y_pred, color = 'violet')
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
url = '/content/insurance.csv'
data = pd.read_csv(url)
data.sample(5)
```

```python
data.info()
```

```python
print(data['sex'].value_counts())
sns.countplot(x='sex', data=data)
print(data['smoker'].value_counts())
sns.countplot(x='smoker',data=data)
```

```python
print(data['region'].value_counts())
sns.countplot(x='region', data = data)
```

```python
corr=data.corr()
sns.heatmap(corr,cmap='Wistia', annot=True)
```

```python
correlation = data.corr()
print(correlation)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['age']), 
          y=data['charges'])
print('Hệ số intercept là: ', model.intercept_)
print('Hệ số coeffient là: ', model.coef_)
```

```python
model.score(X=pd.DataFrame(data['age']), 
          y=data['charges']) 
```

```python
y_pred = model.predict(X = pd.DataFrame(data['age']))
print(y_pred)
```

```python
# vẽ đồ thị 
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9),
          color = 'green')

plt.plot(data['age'], y_pred, color='yellow')
```

```python
# bac 1
model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'],  data['age']**1]).T
y=data['charges']

model.fit(X,y=data['charges'])
print('Intecept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y))
```

```python
import numpy as np
range = np.arange(0, 101, 1.5)
X_range=pd.DataFrame([range,range**1]).T 
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x='age', 
          y = 'charges', figsize = (9,9), color = 'green',
          xlim = (0,100))
plt.plot(range, y_pred, color = 'violet')
```

```python
# bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['age'],  data['age']**1, data['age']**2]).T
y=data['charges']

model.fit(X,y=data['charges'])
print('Intecept của mô hình: ', model.intercept_)
print('Coeffient của mô hình: ', model.coef_)
print('Độ chính xác của mô hình: ', model.score(X,y))
```

```python
import numpy as np
range = np.arange(0, 101, 1.5)
X_range=pd.DataFrame([range,range**1,range**2]).T 
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x='age', 
          y = 'charges', figsize = (9,9), color = 'red',
          xlim = (0,100))
plt.plot(range, y_pred, color = 'violet')
```
